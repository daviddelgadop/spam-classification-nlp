```python
from gensim import corpora

# tokeniser le texte si ce n'est pas encore fait
lemmatized_tokens = texte_lemmatized_spacy_fr.split()

# créer un objet gensim.corpora.Dictionary
dictionary = corpora.Dictionary([lemmatized_tokens])

# convertir lemmatized_tokens en BoW
corpus = [dictionary.doc2bow(lemmatized_tokens)]

print(dictionary)
print("\n--- corpus---")
print(corpus)
```

```python
from gensim.models import LdaModel

# définir le nombre de topics
num_topics = 5

# entrainer le modèle LDA
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, random_state=42, passes=15)

print(f"--- LDA est entrainé avec {num_topics} topics ---")
print(lda_model.print_topics())
```

```python
import pyLDAvis.gensim_models as gensim_models
import pyLDAvis

# Prepare the visualization
lda_display = gensim_models.prepare(lda_model, corpus, dictionary, sort_topics=False)

# Display the visualization
pyLDAvis.display(lda_display)
```
