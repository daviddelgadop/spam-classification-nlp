```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
model_name = "Helsinki-NLP/opus-mt-fr-en"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
text = "j'aime bien le nlp, mais je n'aime pas les embeddings"
inputs = tokenizer(text, return_tensors="pt")
translated = model.generate(**inputs)
print(tokenizer.decode(translated[0], skip_special_tokens=True))
```

```python
from nltk.translate.bleu_score import sentence_bleu

# d√©finir la traduction correcte (ou issue d'un autre mod√®le)
reference = [['i', 'like', 'nlp', 'but', 'i', 'don', 't', 'like', 'embeddings']]

# d√©finir la traduction candidate (celle de notre mod√®le d'interet)
candidate = ['i', 'like', 'nlp', 'but', 'i', 'don', 't', 'like', 'beddings']

# calculer le score BLEU
score = sentence_bleu(reference, candidate)

print(f"BLEU score: {score:.2f}")
```
